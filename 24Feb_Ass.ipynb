```python
def reindex(df):
    df_reset=df.reset_index(drop=True)
    
    new_index_values=[i* 2+1 for i in range(len(df_reset))]
    df_reset.index=new_index_values
    
    return df_reset

new_df=reindex(df)
new_df.head()
```

```python
import pandas as pd
```

```python
data={"date":pd.date_range(start="2024-03-01",end="2024-03-10"),
      "Sales":[100,200,300,130,140,110,170,180,300,400]}
```

```python
df=pd.DataFrame(data)
df.sort_values(by="date",inplace=True)
```

```python
df["MovingAverage"]=df["Sales"].rolling(window=7,min_periods=1).mean()
```

```python
df
```
